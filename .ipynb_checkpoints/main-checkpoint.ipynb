```python
# inputs:   folder name
#           number of choices in each question
#           number of questions
#           model answer text file
folder = '2'
mcq = 2
quesnum = 16
model_answer = '2.txt'

# getting path from folder name
files = os.listdir(folder)
paths = folder + '/' + files[0]

#creating an array of images in that file
images = []
for file in range(len(files)):
    path = folder + '/' + files[file]
    images.append(cv2.cvtColor(io.imread(path), cv2.COLOR_BGR2GRAY))  
    
#getting the model answer as an array
my_data = np.loadtxt(model_answer, delimiter=",", unpack=True)
list2= my_data[:]

#creating an array of warped images
imgs_warp = []
for image in range(len(images)):
    #getting contours and corner points of the question paper
    c = functions.getCVRectContour(images[image])
    p = functions.getCVCornerPoints(c[0])
    
    widthImg = images[image].shape[0]
    heightImg = images[image].shape[1]
    
    #ordering the points to be used in warping
    points = functions.getOrderedCornerPoints(p)
    p2 = np.float32([[0,0], [widthImg,0], [0,heightImg], [widthImg, heightImg]])
    p1 = np.float32(points)
    
    #getting warped image and resizing it to a fixed size 
    img = cv2.getPerspectiveTransform(p1, p2)
    img_warp1 = cv2.warpPerspective(images[image],img, (widthImg, heightImg))
    imgs_warp.append(resize(img_warp1, (800, 1000)))

#getting an array of ID regions    
ID_regs = []
for img_warp in range(len(imgs_warp)):
    #getting ID regions for different sizes of IDs
    if folder == '1':
        ID_regs.append(functions.getIDRegion(imgs_warp[img_warp], index = ID_1[img_warp]))
        rowsnum = 2
    elif folder == '2':
        ID_regs.append(functions.getIDRegion(imgs_warp[img_warp], index = ID_2[img_warp]))
        rowsnum = 4
    elif folder == '3':
        ID_regs.append(functions.getIDRegion(imgs_warp[img_warp], index = ID_3[img_warp]))
        rowsnum = 4
    elif folder == '4':
        rowsnum = 7
        ID_regs.append(functions.getIDRegion(imgs_warp[img_warp], index = ID_4[img_warp]))

#show_images(ID_regs)

#getting an array of individual ID bubbles
IDbubbles = []
for ID_reg in range(len(ID_regs)):
    IDbubbles.append(functions.getbubble(ID_regs[ID_reg], rowsnum))

#getting the ID as a number to be exported to the excel sheet
IDfinal=[]
for IDbubble in range(len(IDbubbles)):
    ID = functions.getID(rowsnum, IDbubbles[IDbubble])
    IDfinal1=[]
    IDfinal1.insert(0,(functions.convert(ID)))
    IDfinal.append(IDfinal1)
    
#print(IDfinal)
#cell_list=IDfinal
#arr = np.array(cell_list)
#dataframe = pd.DataFrame(arr)
#data = dataframe.style.set_properties(align="left")
#data.to_excel(r"book1.xlsx",index=None,header="code")

#getting the answer for each question and the total score and storing them in an array
s = []
q = []
for img_warp in range(len(imgs_warp)):
    score=0
    quesarr=[]
    for i in range (quesnum): 
        Q=functions.getQuestionRegion(imgs_warp[img_warp], index = i,mcq = mcq)
        #show_images([ques])
        Qbubbles = functions.splitQcolumns(Q, mcq)
        #show_images(Qbubbles)
        answer1=functions.getQanswer(mcq,Qbubbles)
        answer=answer1[0]+1
        if(answer==list2[i]):
            score=score+1
            quesarr.insert(i,1)
        else:
            quesarr.insert(i,0)
    #print(quesarr)
    s.insert(img_warp,score)
    q.insert(img_warp,quesarr)


# exporting the data into an excel sheet
data1=IDfinal
data2 = np.array(q)
data3=s

data = [data1, data2, data3]
headers = ['ID', 'Question Grade', 'Total Grade']
df = pd.DataFrame(list(zip(data1, data2, data3)))
df.to_excel(excel_writer = "book1.xlsx")

#df = pd.DataFrame({'ID': data1,
#                   'Questions': data2,
#                  'Score' : data3})
#df.to_excel(r"book1.xlsx",startrow=1,index=False,header=False)

```
